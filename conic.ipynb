```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi

u=0    #x-position of the center
v=0    #y-position of the center
a=10.     #radius on the x-axis
b=5    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
x=np.arange(0,11,1)
y=((a*b)-(b*x))/a

X=(b/a)*(a-x)
Y=(b/a)*(np.sqrt(a**2-x**2))

plt.fill_between(X,Y,0, color='blue', alpha=.2)
plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.plot(x,y,'k--')
plt.grid(color='lightgray',linestyle='--')plt.show()
```
