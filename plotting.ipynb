#2d plot

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,1,10);
y=3*x**2+4;

xp,yp=np.meshgrid(x,y);
plt.quiver(xp,yp,-xp,-yp);
```

#3D plot

```python
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

plt.gca(projection='3d');
x=np.linspace(0,1,10);
y=np.linspace(0,1,10);
z=np.linspace(0,1,10);

xp,yp,zp=np.meshgrid(x,y,z);
plt.quiver(xp,yp,zp,-xp,-yp,-zp)

```

#surface plotting

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
axes = fig.gca(projection='3d')

x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
xv, yv = np.meshgrid(x,y)
z = (xv**2 + yv**2)/2;

axes.plot_surface(xv,yv,z);
```

#contour plot

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

fig=plt.figure(figsize=(8,3));
ax1=fig.add_axes([1,1,1,1]);
#ax2=fig.add_axes([1,1,1,1]);
ax2= fig.gca(projection='3d')

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xv, yv = np.meshgrid(x,y)
z = -np.log(np.sqrt((xv**2 + yv**2)))

ax2.plot_surface(xv,yv,z);
ax1.contour(xv,yv,z,6);
```

#density plots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,100)
y = np.linspace(-1,1,100)
xv, yv = np.meshgrid(x,y)
z = (xv**2 + yv**2)/2

plt.imshow(z, vmin=abs(z).min(), vmax=abs(z).max(),extent=[-1,1,-1,1])
plt.colorbar()
```
